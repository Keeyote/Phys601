# Spiff shoots a pellet

In this imagined problem, Spaceman Spiff (yes, from Calvin and Hobbes) shoots a pellet from a railgun to a target at the front of his spaceship, which is traveling at 0.7c relative to Zephyr, who is observing the target practice.

What do the world lines for Spiff, Zephyr and the pellet look like on a Minkowsky space-time diagram? 

```python
import plotly.graph_objects as go
import plotly.io as pio
from MinkowskiPlot import *
pio.templates.default="plotly_white"

plot = MinkowsiPlot()

spiff = Actor("Spiff",0.,0.,'red')
zephyr = Actor("Zephyr",-0.3,0.,'green')
pellet = Actor("pellet",0.6,0.,'blue')
pellet.PositiveOnly=True

plot.add_actor(spiff)
plot.add_actor(zephyr)
plot.add_actor(pellet)

fig = plot.make_figure()

fig.show()
```
